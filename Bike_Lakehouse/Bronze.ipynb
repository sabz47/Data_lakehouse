#Read From CSV & Write Bronze Table

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} -> bronze.{item['table']}")

    df = ( 
    spark.read
        .option("header","true")
        .option("inferSchema","true")
        .csv(item["path"])
    )
    (
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable(f"bronze.{item['table']}")
    )
```
